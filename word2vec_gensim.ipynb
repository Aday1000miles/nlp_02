```python
#词向量
import pandas as pd
import jieba
from gensim.models.word2vec import Word2Vec
print("\r")
```

```python
# 读入训练集文件
data = pd.read_csv('D:/python/nlp/data/train.csv')
```

```python
# 转字符串数组
corpus = data['comment'].values.astype(str)
```

```python
# 分词，再重组为字符串数组
corpus = [jieba.lcut(corpus[index]
                          .replace("，", "")
                          .replace("!", "")
                          .replace("！", "")
                          .replace("。", "")
                          .replace("~", "")
                          .replace("；", "")
                          .replace("？", "")
                          .replace("?", "")
                          .replace("【", "")
                          .replace("】", "")
                          .replace("#", "")
                        ) for index in range(len(corpus))]
```

```python
# 词向量模型训练
model = Word2Vec(corpus, sg=0, vector_size=300, window=5, min_count=3, workers=4)
```

```python
#模型显示
print('模型参数：',model,'\n')
```

```python
#最匹配
print('最匹配的词是：',model.wv.most_similar(positive=['点赞', '不错'], negative=['难吃']),'\n')
```

```python
#最不匹配
#print('最不匹配的词是：',model.wv.doesnt_match("点赞 好吃 支持 难吃".split()),'\n')
```

```python
#语义相似度
print('相似度为=',model.wv.similarity('推荐','好吃'),'\n')
```

```python
#坐标返回
print(model.wv.__getitem__('地道'))
```

1、使用Skip-Gram训练Word2Vec模型（注意代码示例里使用的是CBOW）

```python
# 词向量模型训练
model = Word2Vec(corpus, sg=1, vector_size=300, window=5, min_count=3, workers=4)
```

```python
print('模型参数：',model,'\n')
```

2、输出“环境”的词向量及其形状

```python
print(model.wv.__getitem__('环境'))
```

3、输出与“好吃”语义最接近的3个词

```python
word = "好吃"
similar_words = model.wv.most_similar(word, topn=3)
print(f"与'{word}'\最相似的3个词：{similar_words}")
```

```python
word = "好吃"
if word in model.wv:
    similar_words = model.wv.most_similar(word, topn=3)
    print(f"与'{word}'最相似的3个词：")
    for w, score in similar_words:
        print(f"{w}相似度: \t{score:.4f}")
else:
    print(f"词 '{word}' 不在词汇表中\t")
```

4、计算“好吃”和“美味”的相似度、“好吃”和“蟑螂”的相似度

```python
list = ["好吃", "美味", "蟑螂"]
w1 = list[0]
for i in range(1, 3, 1):
    w2 = list[i]
    print('相似度为 =',model.wv.similarity(w1, w2))
```

5、执行向量运算“餐厅+聚会-安静=？”，输出1个最相关结果（类比“王子-男+女=公主”）

```python
list = [["餐厅", "聚会", "安静"],
         ["王子", "女", "男"]]
for row in list:
    w, w_p, w_n = row  # 解包每行
    # 检查词是否在词向量表中
    if all(word in model.wv for word in [w, w_p, w_n]):
        result = model.wv.most_similar(positive=[w, w_p], negative=[w_n], topn=1)
        print(f"{w} + {w_p} - {w_n} => {result[0][0]}, 相似度: {result[0][1]:.4f}")
    else:
        print(f"{row}中有词不在词汇表中")
```
