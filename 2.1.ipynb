# 结巴分词实践

## 结巴分词的三种模式

### 全模式

```python
import jieba
```

```python
seg_list_default = jieba.cut("我来到北京清华大学", cut_all=False)
print("Default Mode: " + "/ ".join(seg_list_default))  # 精确模式
```

### 精确模式

```python
seg_list_full = jieba.cut("我来到北京清华大学", cut_all=True)
print("Full Mode: " + "/ ".join(seg_list_full))  # 全模式
```

### 搜索引擎模式

```python
seg_list_search = jieba.cut_for_search("小明硕士毕业于中国科学院计算所，后在日本京都大学深造")  # 搜索引擎模式
print(", ".join(seg_list_search))
```

### 载入词典

```python
jieba.load_userdict("D:/python/nlp/data/userdict.txt")
seg_list = jieba.cut("陶量即将是自然语言处理方面的高手", cut_all = False, HMM = False)
print("Default Mode: " + "/ ".join(seg_list))
```
